## 3. Feature Extraction using CountVectorizer
This section transforms the raw text data into a numerical format suitable for machine learning algorithms. We use `CountVectorizer` to create a bag-of-words representation, which is a common and effective method for converting textual data into numerical feature vectors that can be used by similarity algorithms like cosine similarity. This process helps us quantify the presence of words in each song's lyrics.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=34000, stop_words='english')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  a=[]
  for i in text.split():
    a.append(ps.stem(i))

  return ' '.join(a)
```

```python
stem('loved to hell')
```

```python
song['text'] = song['text'].apply(stem)
```

```python
song['text'][0]
```

## 5. Similarity Calculation and Recommendation System Implementation
This is the core of our recommendation system. We calculate the cosine similarity between all song lyrical vectors and implement a function to recommend songs based on a given input song.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(cs[0],)),reverse=True,key=lambda x:x[1])[1:6]
```

```python
song[song['song'] =='Bang'].index[0]
```

```python
song.iloc[0].song
```

```python
def Recommend(songs):
  song_index = song[song['song'] == songs].index[0]
  distance = cs[song_index]
  top_song = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  for i in top_song:
    print(song.iloc[i[0]].song)
```

# RESULT :

```python
Recommend('Bang-A-Boomerang')
```
